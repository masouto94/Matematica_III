## Operaciones con eventos

El espacio muestral, los eventos y sus relaciones se pueden representar convenientemente mediante los diagramas de Venn.  
En estos diagramas el espacio muestral  est representado por el conjunto de puntos, que se encuentran en el interior de una determinada figura geomtrica, por lo general un rectngulo.  
Los eventos $E_{i}$ se representan como figuras de menor o igual tamao que .  
Dado los eventos $E_{1}$, $E_{2}$ y el espacio muestral  se pueden realizar las siguientes operaciones:  

- La diferencia de dos eventos es el subconjunto del espacio muestral que contiene a los elementos de $E_{1}$ que no se encuentran en $E_{2}$:

$$E_{1}  E_{2} = \{x/x    E_{1}  x \not  E_{2}\}$$  

Que dicho en criollo sera: X tal que sea perteneciente a E1 y no pertenezca a E2

  

Si $E_{1} = \{a, b, c, d\}$ y $E_{2} = \{a, e, i, o\}$ entonces:  

$$E_{1}  E_{2} = \{b, c, d\};$$  

- El complemento de un evento es el subconjunto de
todos los elementos del espacio muestral , que no
pertenecen a E:
$$E^c =   E = \{x/x    x \not  E\}$$  

  


- La unin de dos eventos es el subconjunto del espacio muestral que contiene a los elementos de $E_{1}$ y $E_{2}$, pero sin repetir los elementos:  

$$ E_{1}  E_{2} = \{x/x  E_{1}  E_{2}\} $$
Si $E_{1} = \{a, b, c, d\}$ y $E_{2} = \{a, e, i, o\} $ entonces:  

$$ E_{1}  E_{2} = \{a, b, c, d, e, i, o\}; $$  



- La interseccin de dos eventos es el subconjunto del
espacio muestral que contiene a aquellos elementos
que pertenecen tanto a $E_{1}$ y a $E_{2}$:  

$$E1  E2 = \{x/x  E_{1}  E_{2} \}$$  



Si $E_{1} = \{a, b, c, d\}$ y $E_{2} = \{a, e, i, o\}$ entonces:  

$$E_{1}  E_{2} = \{a\};$$  

Las siguientes definiciones son aplicables a los eventos:
- $E_{1}$ y $E_{2}$ son mutuamente excluyentes si no pueden
ocurrir al mismo tiempo. Si uno de los eventos ocurre,
el otro no puede ocurrir, y viceversa. En el lanzamiento de una moneda solo puede salir cara o cruz,
pero no ambas;
- Un conjunto de eventos son colectivamente exhaustivo si su unin da como resultado el espacio muestral, $E_{1}$  $E_{2}$      $E_{n}$ = .
- $E_{1}$ y $E_{2}$ son independientes si al ocurrir uno no afecta
la ocurrencia del otro.

## Propiedades de las operaciones con eventos  

1. Propiedad conmutativa:  
    - $E_{1}  E_{2} = E_{2}  E_{1}$
    - $E_{1}  E_{2} = E_{2}  E_{1}$
1. Propiedad distributiva:
    - $E_{1}  (E_{2}  E_{3}) = (E_{1}  E_{2})  (E_{1}  E_{3})$
    - $E_{1}  (E_{2}  E_{3}) = (E_{1}  E_{2})  (E_{1}  E_{3})$
1. Propiedades de la diferencia de conjuntos:
    - $E_{1}  E_{2} = E_{1}  E^c_{2}$
    - $(E_{1}  E_{2})  E_{3} = (E_{1}  E_{3})  (E_{2}  E_{3})$
    - $E_{1}  (E_{2}  E_{3}) = (E_{1}  E_{2})  (E_{1}  E_{3})$
1. Las leyes de Morgan:
    - $(E_{1}  E_{2})^c = E^c_{1}  E^c_{2}$
    - $(E_{1}  E_{2})^c = E^c_{1}  E^c_{2}$

## La probabilidad condicional  

La probabilidad condicional es la probabilidad de que ocurra un evento cuando se sabe que ocurre otro evento especfico.  
La notacin de la probabilidad condicional es:  
$$p(E_{1}|E_{2})$$  

y se lee la probabilidad del evento 1 dado el evento 2, es decir que ocurre el evento 2 s o s.

En el ejemplo anterior la probabilidad de seleccionar un
estudiante que estudie fsica sabiendo que estudia qumica
es:  

$$p (E_{1}|E_{2}) = 2/4 = 0.5$$

debido a que los estudiantes que estudian qumica son 4
y de entre ellos solo 2 estudian fsica. Por otro lado la
probabilidad de seleccionar un joven que estudie qumica

## La probabilidad conjunta

La probabilidad conjunta es la probabilidad de que dos
eventos sucedan al mismo tiempo.  
Matemticamente se denota por $p(E_{1}  E_{2})$. En el ejemplo anterior la probabilidad
de seleccionar un joven que estudie ambas asignaturas es:  
$$p(E_{1}  E_{2}) = 2/7 = 0.286$$  

La probabilidad conjunta de que ocurran dos eventos se
puede calcular a partir de las probabilidades condicionales
y marginales de la siguiente forma:  

$$p(E_{1}  E_{2}) = p(E_{1}|E_{2}) p(E_{2})$$  

o tambin como:  

$$p(E_{1}  E_{2}) = p(E_{2}|E_{1}) p(E_{1})$$

## El teorema de la probabilidad total  
Sea que el espacio muestral se puede descomponer en distintos eventos mutuamente excluyentes $E_{i}$ y adems en el espacio muestral est definido un evento A como lo muestra la siguiente figura, entonces la cardinalidad del evento A est dada por:  

$$N(A) = N(E_{1}  A) + N(E_{2}  A) + N(E_{3}  A)$$  

y su respectiva probabilidad esta dada por:  

$$p(A) = \frac{N(E_{1}  A) + N(E_{2}  A) + N(E_{3}  A)}{N()}$$


A la probabilidad de ocurrencia del evento A considerando todos los dems eventos se le denomina probabilidad total y en trminos generales se escribe como:  

$$p(A) = \displaystyle\sum_{i=1}^n p(E_{i}  A)$$

La probabilidad total relaciona las probabilidades marginales
con las probabilidades condicionales.  

$$p(E_{i}  A) = p(A|E_{i}) p(E_{i})$$

Por lo tanto,  
$$p(A) = \displaystyle\sum_{i=1}^n p(A|E_{i}) p(E_{i})$$

## El teorema de Bayes  

Thomas Bayes propuso un mtodo para calcular la probabilidad condicional de un evento $E$ respecto a otro evento
$F$ si se conoce las probabilidades de los eventos y la probabilidad condicional de $F$ sabiendo que ha ocurrido $E$.
De la probabilidad conjunta de dos eventos se tiene:  

$$p(E  F) = p(E|F) p(F)$$
y
$$p(E  F) = p(F|E) p(E)$$
Por lo tanto,
$$p(E|F) p(F) = p(F|E) p(E)$$
y de aqu se obtiene:
$$p(E|F) = \frac{p(F|E) p(E)}{p(F)}$$

En trminos generales, dado un nmero de eventos que forman un grupo exhaustivo y considerando el teorema de la probabilidad total, el teorema de Bayes se escribe de la siguiente forma:  

$$p(E_{i}|F) = \frac{p(F|E_{i}) p(E_{i})}{\displaystyle\sum_{i=1}^n p(F|E_{i})p(E_{i})}$$

### Ejemplo  

Una fbrica de papas fritas (chips) compra papas a dos proveedores.

*Resolucin*:  

Definamos los eventos:
E1 = {comprada al primer proveedor};
E2 = {comprada al segundo proveedor};
F = {la papa es de rechazo}.
Las probabilidades marginales se pueden calcular a partir de la cuota de cada proveedor. Si se supone que se
compra $n$ kilogramos de papa
$$n = n_{1}+n_{2}$$

$n_{1}$ - es la cantidad de kilogramos comprados al primer
proveedor;  
$n_{2}$ - es la cantidad de kilogramos comprados al segundo
proveedor;  

$$p(E_{1}) = \frac{n_{1}}{n_{1}+\frac{1}{5}n_{1}} = \frac{5}{6} = 0.833$$

$$p(E_{2}) = 1  p(E_{1}) = \frac{1}{6} = 0.167$$

Las probabilidades condicionales de seleccionar una papa de rechazo de un proveedor determinado son:

El teorema de Bayes generalmente se interpreta en trminos de causa - efecto. Los eventos $E_{i}$ se consideran las
causas y el evento $F$; el efecto. Un ejemplo de causa - efecto puede ser el hecho de que un auto no enciende.  
Se puede sealar las siguientes causas:  

- la batera est agotada;
- se agot el combustible;
- hay un desperfecto elctrico, entre otros.  

Se intenta encender el auto y este no enciende.  
Cul es la probabilidad de que el auto no encienda debido a que la batera est agotada?  
El teorema de Bayes permite evaluar la probabilidad de que la batera sea la causa por la cul no enciende el auto.  

El teorema de Bayes tambin se interpreta en trminos de hiptesis.  
En este caso los eventos $E_{i}$ son las hiptesis.  
Por ejemplo, si se considera una enfermedad que afecta a una poblacin y se toma una persona al azar se puede considerar la hiptesis de que est enfermo o su complemento de que se encuentre sano.  
En ausencia de evidencias esta probabilidad se toma en funcin de la prevalencia.  
Es decir en base a las estadsticas de la poblacin.  

La determinacin de si una persona est enferma o no se afecta drsticamente con la aportacin de evidencias, por ejemplo, la realizacin de una prueba de laboratorio.  
Sin embargo, las pruebas de laboratorio no son perfectas y a pesar de que la prueba d positivo no se tiene la certeza de que la persona est enferma.  

La informacin que se tiene antes de realizar la prueba se llama a priori. Como resultado de la prueba se decide si la persona seleccionada est enferma con una probabilidad a posteriori. Obviamente esta probabilidad est mejorada dada la aportacin de evidencia (la prueba).  
Con el objetivo de incrementar la certeza en la hiptesis se puede realizar otra prueba.  
En este caso la probabilidad a posteriori pasa a ser a priori.  
La probabilidad de que la persona est enferma dado que la prueba da positivo de acuerdo al teorema de Bayes es:  

$$p(E|F) = p(F|E) p(E)$$  

- $p(E|F)$ es la probabilidad a posteriori;
- $p(E)$ es la probabilidad a priori;
- $p(F)$ es la probabilidad total de que la evidencia sea positiva y
- $p(F|E)$ es la probabilidad de que la prueba de un verdadero positivo

### Ejemplo en Python  

Escriba un cdigo en Python para determinar el nmero de pruebas necesarias para tener una certeza superior al 95%   
de que una persona se encuentre afectada por una enfermedad si se sabe que la sensibilidad y la especificidad  
de las pruebas son 0.9 y 0.95 respectivamente y la prevalencia de la enfermedad es del 1%.

```python
import numpy as np
import matplotlib.pyplot as plt
prevalencia = 0.01
sensibilidad = 0.9
especificidad = 0.95
p = prevalencia
P = np.array([])
while p<0.95:
    ptotal = sensibilidad * p + (1 - especificidad) * (1 - p)
    p = sensibilidad * p / ptotal
    P = np.append(P, p)
plt.plot(range(3), P)
plt.xlabel("Nmero de pruebas", fontsize=14)
plt.ylabel("Certeza de estar enfermo", fontsize=14)
plt.locator_params(axis="x",nbins=3)
plt.grid(axis="y")
plt.show()
```
