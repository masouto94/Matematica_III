# Probabilidad condicional y teorema de Bayes

## Operaciones con eventos

El espacio muestral, los eventos y sus relaciones se pueden representar convenientemente mediante los diagramas de Venn.  
En estos diagramas el espacio muestral Ω está representado por el conjunto de puntos, que se encuentran en el interior de una determinada figura geométrica, por lo general un rectángulo.  
Los eventos $E_{i}$ se representan como figuras de menor o igual tamaño que Ω.  
Dado los eventos $E_{1}$, $E_{2}$ y el espacio muestral Ω se pueden realizar las siguientes operaciones:  

- La diferencia de dos eventos es el subconjunto del espacio muestral que contiene a los elementos de $E_{1}$ que no se encuentran en $E_{2}$:

$$E_{1} − E_{2} = \{x/x  ∈  E_{1} ∧ x \not ∈ E_{2}\}$$  

Que dicho en criollo sería: X tal que sea perteneciente a E1 y no pertenezca a E2

  

Si $E_{1} = \{a, b, c, d\}$ y $E_{2} = \{a, e, i, o\}$ entonces:  

$$E_{1} − E_{2} = \{b, c, d\};$$  

- El complemento de un evento es el subconjunto de
todos los elementos del espacio muestral Ω, que no
pertenecen a E:
$$E^c = Ω − E = \{x/x ∈ Ω ∧ x \not ∈ E\}$$  

  


- La unión de dos eventos es el subconjunto del espacio muestral que contiene a los elementos de $E_{1}$ y $E_{2}$, pero sin repetir los elementos:  

$$ E_{1} ∪ E_{2} = \{x/x ∈ E_{1} ∨ E_{2}\} $$
Si $E_{1} = \{a, b, c, d\}$ y $E_{2} = \{a, e, i, o\} $ entonces:  

$$ E_{1} ∪ E_{2} = \{a, b, c, d, e, i, o\}; $$  



- La intersección de dos eventos es el subconjunto del
espacio muestral que contiene a aquellos elementos
que pertenecen tanto a $E_{1}$ y a $E_{2}$:  

$$E1 ∩ E2 = \{x/x ∈ E_{1} ∧ E_{2} \}$$  



Si $E_{1} = \{a, b, c, d\}$ y $E_{2} = \{a, e, i, o\}$ entonces:  

$$E_{1} ∩ E_{2} = \{a\};$$  

Las siguientes definiciones son aplicables a los eventos:
- $E_{1}$ y $E_{2}$ son mutuamente excluyentes si no pueden
ocurrir al mismo tiempo. Si uno de los eventos ocurre,
el otro no puede ocurrir, y viceversa. En el lanzamiento de una moneda solo puede salir cara o cruz,
pero no ambas;
- Un conjunto de eventos son colectivamente exhaustivo si su unión da como resultado el espacio muestral, $E_{1}$ ∪ $E_{2}$ ∪ · · · ∪ $E_{n}$ = Ω.
- $E_{1}$ y $E_{2}$ son independientes si al ocurrir uno no afecta
la ocurrencia del otro.

  

$E_{1}$ y $E_{3}$ son mutuamente excluyentes, pero no
$E_{1}$ y $E_{2}$.

## Propiedades de las operaciones con eventos  

1. Propiedad conmutativa:  
    - $E_{1} ∪ E_{2} = E_{2} ∪ E_{1}$
    - $E_{1} ∩ E_{2} = E_{2} ∩ E_{1}$
1. Propiedad distributiva:
    - $E_{1} ∩ (E_{2} ∪ E_{3}) = (E_{1} ∩ E_{2}) ∪ (E_{1} ∩ E_{3})$
    - $E_{1} ∪ (E_{2} ∩ E_{3}) = (E_{1} ∪ E_{2}) ∩ (E_{1} ∪ E_{3})$
1. Propiedades de la diferencia de conjuntos:
    - $E_{1} − E_{2} = E_{1} ∩ E^c_{2}$
    - $(E_{1} ∩ E_{2}) − E_{3} = (E_{1} − E_{3}) ∩ (E_{2} − E_{3})$
    - $E_{1} ∩ (E_{2} − E_{3}) = (E_{1} ∩ E_{2}) − (E_{1} ∩ E_{3})$
1. Las leyes de Morgan:
    - $(E_{1} ∪ E_{2})^c = E^c_{1} ∩ E^c_{2}$
    - $(E_{1} ∩ E_{2})^c = E^c_{1} ∪ E^c_{2}$

un solo evento independientemente de otros eventos.

## La probabilidad condicional  

La probabilidad condicional es la probabilidad de que ocurra un evento cuando se sabe que ocurre otro evento específico.  
La notación de la probabilidad condicional es:  
$$p(E_{1}|E_{2})$$  

y se lee la probabilidad del evento 1 dado el evento 2, es decir que ocurre el evento 2 sí o sí.

En el ejemplo anterior la probabilidad de seleccionar un
estudiante que estudie física sabiendo que estudia química
es:  

$$p (E_{1}|E_{2}) = 2/4 = 0.5$$

debido a que los estudiantes que estudian química son 4
y de entre ellos solo 2 estudian física. Por otro lado la
probabilidad de seleccionar un joven que estudie química
dado que estudia física es:

$$p(E_{2}|E_{1}) = 2/5 = 0.4$$

## La probabilidad conjunta

La probabilidad conjunta es la probabilidad de que dos
eventos sucedan al mismo tiempo.  
Matemáticamente se denota por $p(E_{1} ∩ E_{2})$. En el ejemplo anterior la probabilidad
de seleccionar un joven que estudie ambas asignaturas es:  
$$p(E_{1} ∩ E_{2}) = 2/7 = 0.286$$  

La probabilidad conjunta de que ocurran dos eventos se
puede calcular a partir de las probabilidades condicionales
y marginales de la siguiente forma:  

$$p(E_{1} ∩ E_{2}) = p(E_{1}|E_{2}) p(E_{2})$$  

o también como:  

$$p(E_{1} ∩ E_{2}) = p(E_{2}|E_{1}) p(E_{1})$$

## El teorema de la probabilidad total  
Sea que el espacio muestral se puede descomponer en distintos eventos mutuamente excluyentes $E_{i}$ y además en el espacio muestral está definido un evento A como lo muestra la siguiente figura, entonces la cardinalidad del evento A está dada por:  

$$N(A) = N(E_{1} ∩ A) + N(E_{2} ∩ A) + N(E_{3} ∩ A)$$  

y su respectiva probabilidad esta dada por:  

$$p(A) = \frac{N(E_{1} ∩ A) + N(E_{2} ∩ A) + N(E_{3} ∩ A)}{N(Ω)}$$


A la probabilidad de ocurrencia del evento A considerando todos los demás eventos se le denomina probabilidad total y en términos generales se escribe como:  

$$p(A) = \displaystyle\sum_{i=1}^n p(E_{i} ∩ A)$$

La probabilidad total relaciona las probabilidades marginales
con las probabilidades condicionales.  

$$p(E_{i} ∩ A) = p(A|E_{i}) p(E_{i})$$

Por lo tanto,  
$$p(A) = \displaystyle\sum_{i=1}^n p(A|E_{i}) p(E_{i})$$

### Ejemplo
Se inspecciona tres grupos de puertas y se encuentra que:

Si se elige una puerta al azar, determine la probabilidad de que esta sea defectuosa.

Definamos los eventos:

El número total de puertas inspeccionadas es:
$$N = 380 + 270 + 350 = 1000$$  

Las probabilidades marginales son:  

$$p (E1) = \frac{380}{1000} = 0.380$$
$$p (E2) = \frac{270}{1000} = 0.270$$
$$p (E3) = \frac{350}{1000} = 0.350$$

## El teorema de Bayes  

Thomas Bayes propuso un método para calcular la probabilidad condicional de un evento $E$ respecto a otro evento
$F$ si se conoce las probabilidades de los eventos y la probabilidad condicional de $F$ sabiendo que ha ocurrido $E$.
De la probabilidad conjunta de dos eventos se tiene:  

$$p(E ∩ F) = p(E|F) p(F)$$
y
$$p(E ∩ F) = p(F|E) p(E)$$
Por lo tanto,
$$p(E|F) p(F) = p(F|E) p(E)$$
y de aquí se obtiene:
$$p(E|F) = \frac{p(F|E) p(E)}{p(F)}$$

En términos generales, dado un número de eventos que forman un grupo exhaustivo y considerando el teorema de la probabilidad total, el teorema de Bayes se escribe de la siguiente forma:  

$$p(E_{i}|F) = \frac{p(F|E_{i}) p(E_{i})}{\displaystyle\sum_{i=1}^n p(F|E_{i})p(E_{i})}$$

### Ejemplo  

Una fábrica de papas fritas (chips) compra papas a dos proveedores.  
Las papas deben tener un tamaño mínimo caso contrario son consideradas de rechazo.  
En un muestreo se determina que el porcentaje de papas rechazadas del primer proveedor es igual a 20%, mientras que del segundo proveedor es el 10%.  
Si se sabe que al primer proveedor se le compra cinco veces más que al segundo, determine la probabilidad de que al encontrar una papa de rechazo en la línea de producción esta sea del segundo proveedor.

*Resolución*:  

Definamos los eventos:
E1 = {comprada al primer proveedor};
E2 = {comprada al segundo proveedor};
F = {la papa es de rechazo}.
Las probabilidades marginales se pueden calcular a partir de la cuota de cada proveedor. Si se supone que se
compra $n$ kilogramos de papa
$$n = n_{1}+n_{2}$$

$n_{1}$ - es la cantidad de kilogramos comprados al primer
proveedor;  
$n_{2}$ - es la cantidad de kilogramos comprados al segundo
proveedor;  

$$p(E_{1}) = \frac{n_{1}}{n_{1}+\frac{1}{5}n_{1}} = \frac{5}{6} = 0.833$$

$$p(E_{2}) = 1 − p(E_{1}) = \frac{1}{6} = 0.167$$

Las probabilidades condicionales de seleccionar una papa de rechazo de un proveedor determinado son:  

$$p(F|E_{1}) = 0.2$$
$$p(F|E_{2}) = 0.1$$

La probabilidad total de seleccionar una papa de rechazo es:  

$$p(F) = \displaystyle\sum_{i=1}^n p(F|E_{i}) p(E_{i})$$
$$p(F) = 0.833 × 0.2 + 0.167 × 0.1 = 0.183$$

De acuerdo al teorema de Bayes:  

$$p(E_{2}|F) = \frac{p(F|E_{2})p(E_{2})}{p(F)}$$

$$p(E_{2}|F) = \frac{0.1 × 0.167}{0.183} = 0.091$$

El teorema de Bayes generalmente se interpreta en términos de causa - efecto. Los eventos $E_{i}$ se consideran las
causas y el evento $F$; el efecto. Un ejemplo de causa - efecto puede ser el hecho de que un auto no enciende.  
Se puede señalar las siguientes causas:  

- la batería está agotada;
- se agotó el combustible;
- hay un desperfecto eléctrico, entre otros.  

Se intenta encender el auto y este no enciende.  
¿Cuál es la probabilidad de que el auto no encienda debido a que la batería está agotada?  
El teorema de Bayes permite evaluar la probabilidad de que la batería sea la causa por la cuál no enciende el auto.  

El teorema de Bayes también se interpreta en términos de hipótesis.  
En este caso los eventos $E_{i}$ son las hipótesis.  
Por ejemplo, si se considera una enfermedad que afecta a una población y se toma una persona al azar se puede considerar la hipótesis de que esté enfermo o su complemento de que se encuentre sano.  
En ausencia de evidencias esta probabilidad se toma en función de la prevalencia.  
Es decir en base a las estadísticas de la población.  

La determinación de si una persona está enferma o no se afecta drásticamente con la aportación de evidencias, por ejemplo, la realización de una prueba de laboratorio.  
Sin embargo, las pruebas de laboratorio no son perfectas y a pesar de que la prueba dé positivo no se tiene la certeza de que la persona esté enferma.  

La información que se tiene antes de realizar la prueba se llama ’a priori’. Como resultado de la prueba se decide si la persona seleccionada está enferma con una probabilidad ’a posteriori’. Obviamente esta probabilidad está mejorada dada la aportación de evidencia (la prueba).  
Con el objetivo de incrementar la certeza en la hipótesis se puede realizar otra prueba.  
En este caso la probabilidad ’a posteriori’ pasa a ser ’a priori’.  
La probabilidad de que la persona esté enferma dado que la prueba da positivo de acuerdo al teorema de Bayes es:  

$$p(E|F) = p(F|E) p(E)$$  

- $p(E|F)$ es la probabilidad ’a posteriori’;
- $p(E)$ es la probabilidad ’a priori’;
- $p(F)$ es la probabilidad total de que la evidencia sea positiva y
- $p(F|E)$ es la probabilidad de que la prueba de un verdadero positivo

### Ejemplo  
Una enfermedad afecta al 1% de la población.  
Las pruebas identifican correctamente al 90% de los casos positivos e identifican incorrectamente al 5% de los negativos.  
Se realizan varios test a una persona. ¿Cuál es la certeza de que esta persona esté enferma después de cada test si estos dan positivo?  

*Resolución*:
De acuerdo al enunciado la prevalencia de la enfermedad es del 1%.  
La sensibilidad (verdadero positivo) es igual a 0.9 y la especificidad (verdadero negativo) es igual a 0.95.  
Antes de realizar el primer test la certeza de que una persona esté enferma es igual a la prevalencia.  
Para calcular la probabilidad después del primer test definamos los eventos:  
E = {está enfermo};  
F = {el test da positivo}.  
Supongamos que se observan 10000 personas para completar el cuadro de datos.  
El número de afectados entre la población es:  

$$N(E) = 0.01 × 10000 = 100$$

El número de casos positivos detectados correctamente por los test es:  

$$N(E ∩ F) = 0.9 × 100 = 90$$  

El número de casos negativos detectados incorrectamente por los tests es:  

$$N(E^c ∩ F) = 0.05 × 9900 = 495$$  

  

A partir del cuadro la certeza de estar enfermo si el test da positivo es:  

$$p(E|F) = \frac{90}{585} = 0.154$$


Para calcular la probabilidad de estar enfermo después del segundo test se debe considerar que el primer test dio positivo y por lo tanto la población analizada
es ahora: $N(F) = 585$.  

  

A partir del cuadro la certeza de estar enfermo si el segundo test da positivo es:  

$$p(E|F) = \frac{81}{105.75} = 0.766$$

Después del tercer test la población es $N (F) =105.75$  

  

A partir del cuadro la certeza de estar enfermo si el tercer test da positivo es:  

$$p(E|F) = \frac{72.9}{74.14} = 0.983$$  

Ahora verificaremos los cálculos utilizando el teorema de Bayes:  

$$p(E|F) = \frac{p(F|E)p(E)}{p(F)}$$

Para el primer test:

$$p(F|E) = 0.9$$
$$p(E) = 0.01$$
$$p(F) = p(F|E) p(E) + p(F|E^c) p(E^c)$$
$$p(F) = 0.9 × 0.01 + 0.05 × (1 − 0.01) = 0.0585$$

$$p(E|F) = \frac{0.9 × 0.01}{0.0585} = 0.154$$

Para el segundo test:  

$$p(F|E) = 0.9$$
$$p(E) = 0.154$$
$$p(F) = 0.9 × 0.154 + 0.05 × (1 − 0.154) = 0.1809$$

$$p(E|F) = \frac{0.9 × 0.154}{0.1809} = 0.766$$  

Para el tercer test:
$$p (F|E) = 0.9$$
$$p (E) = 0.766$$
$$p (F) = 0.9 × 0.766 + 0.05 × (1 − 0.766) = 0.701$$

$$p(E|F) = \frac{0.9 × 0.766}{0.701} = 0.983$$

### Ejemplo en Python  

Escriba un código en Python para determinar el número de pruebas necesarias para tener una certeza superior al 95%   
de que una persona se encuentre afectada por una enfermedad si se sabe que la sensibilidad y la especificidad  
de las pruebas son 0.9 y 0.95 respectivamente y la prevalencia de la enfermedad es del 1%.

```python
import numpy as np
import matplotlib.pyplot as plt
prevalencia = 0.01
sensibilidad = 0.9
especificidad = 0.95
p = prevalencia
P = np.array([])
while p<0.95:
    ptotal = sensibilidad * p + (1 - especificidad) * (1 - p)
    p = sensibilidad * p / ptotal
    P = np.append(P, p)
plt.plot(range(3), P)
plt.xlabel("Número de pruebas", fontsize=14)
plt.ylabel("Certeza de estar enfermo", fontsize=14)
plt.locator_params(axis="x",nbins=3)
plt.grid(axis="y")
plt.show()
```
