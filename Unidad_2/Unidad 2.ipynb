## Teoría de números  


La teoría de números es un campo de las matemáticas que reflexiona sobre la naturaleza y las propiedades de los números, desde los sistemas numéricos antiguos hasta los conceptos modernos. Aquí hay un resumen de los sistemas numéricos principales y sus características:

1. **Números Naturales** $$\Bbb{N}$$  
Los números naturales son los primeros y más básicos: 1, 2, 3, etc.  
Son solo positivos y representan conteos o cantidades enteras no fraccionarias.  
Históricamente, incluso las cavernas muestran marcas de conteo.  

2. **Números Enteros** $$\Bbb{Z}$$  
Incluyen los números naturales, el cero y los números negativos.  
Los números enteros amplían nuestro sistema numérico para incluir conceptos como deudas o temperaturas bajo cero.  
Aunque los números negativos fueron inicialmente desconfiados, su utilidad, especialmente en finanzas y matemáticas avanzadas, los ha hecho indispensables.  

3. **Números Racionales** $$\Bbb{Q}$$  
Son aquellos que pueden expresarse como el cociente de dos enteros, como 2/3 o 6.87 (que sería 687/100).  
Los racionales incluyen fracciones y decimales finitos, útiles para medir cantidades no enteras, como partes de un litro de leche o distancias parciales.

4. **Números Irracionales** $$\Bbb{I}$$  
Estos no pueden expresarse como fracciones y tienen decimales infinitos no repetitivos, como π, la raíz cuadrada de 2, y el número *e*.  
La existencia de números irracionales desafió las creencias matemáticas antiguas, como las de los seguidores de Pitágoras, llevando a importantes avances en matemáticas.  

5. **Números Reales** $$\Bbb{R}$$  
Los números reales combinan racionales e irracionales, cubriendo prácticamente cualquier número que se encuentre en un contexto de ciencia de datos, incluyendo todos los decimales con los que se trabaja típicamente.  

6. **Números Complejos e Imaginarios** $$\Bbb{C}$$  
Aparecen al tomar raíces cuadradas de números negativos.  
Aunque su aplicación es limitada en ciencia de datos básica, juegan un rol en casos más avanzados, como la descomposición de matrices.  

## Orden de operaciones  
El orden de operaciones es fundamental para resolver expresiones matemáticas correctamente.  

Este principio establece que las operaciones deben realizarse en el siguiente orden:  
1. Lo que está dentro de los paréntesis  
2. Los exponentes  
3. Multiplicación y división (que son intercambiables en orden según la secuencia de izquierda a derecha de la operación)  
4. Suma y resta.  

Veamos un ejemplo práctico de cómo aplicar el orden de operaciones:  

$$2 \times \frac {(3+2)^2} {5} - 4$$  

$$2 \times \frac {(5)^2} {5} - 4$$  

$$2 \times \frac {25} {5} - 4$$  

$$2 \times 5 - 4$$  

$$10 - 4 = 6$$  

Como veremos en el ejemplo que viene, **Python** lo expresará como `6.0`.  


```python
my_value = 2 * (3 + 2)**2 / 5 - 4

print(my_value)
```

Usar paréntesis en código no solo aclara el orden de operaciones para evitar confusiones, sino que también sirve como referencia visual durante modificaciones futuras, reduciendo el riesgo de errores.  
Es una práctica recomendada para mantener el código legible y seguro.  

```python
my_value = 2 * ((3 + 2)**2 / 5) - 4

print(my_value)
```

## Funciones
Las funciones son expresiones que definen relaciones entre dos o más variables. Más específicamente, una función toma variables de entrada (también llamadas variables de dominio o variables independientes), las introduce en una expresión, y luego produce una variable de salida (también llamada variable dependiente).

Por ejemplo, tomemos esta función lineal
$$ y = 2x + 1 $$

Para un valor de *x* dado, resolvemos la expresión para encontrar *y*. Ejemplo: cuando *x = 1*, entonces *y = 3*

Las funciones son útiles porque describen de manera predecible cómo se relacionan diferentes variables, como por ejemplo cuántos incendios (*y*) podríamos esperar a una cierta temperatura (*x*). Más adelante, utilizaremos funciones lineales para realizar análisis de regresión lineal.

También es usual ver las funciones de esta forma:

$$ f(x) = 2x + 1 $$

Ahora veremos cómo hacerlo en Python

```python
def f(x):
    return 2 * x + 1

x_values = [0, 1, 2, 3]
for x in x_values:
    y = f(x)
    print(y)
```

## Sumatorias
Una suma se expresa como un símbolo sigma **Σ** y consiste en sumar varios elementos.

Por ejemplo, si queremos sumar los números del 1 al 5, multiplicando por 2 cada uno y sumando, la expresión sería la siguiente:

$$ \displaystyle\sum_{i=1}^5 2{i} = (2) 1 + (2) 2 + (2) 3 + (2) 4 + (2) 5 = 30$$

Para realizarlo en Python haremos lo siguiente:

```python
summation = sum(2*i for i in range(1,6))
print(summation)
```

Observar que *i* es una variable de marcador que representa cada valor de índice consecutivo que estamos recorriendo en el bucle. En este caso, multiplicamos cada valor de *i* por 2 y luego sumamos todos los resultados. 
Al iterar datos, es común encontrarse con variables como $x_{i}$, las cuales indican un elemento en una colección en el índice *i*.

<div class="alert alert-block alert-info">
<b>Nota:</b> la función range() en Python es exclusiva en el límite final, lo que significa que si invocas range(1,4), iterará los números 1, 2 y 3. Excluye el 4 como límite superior.
</div>

También es común ver que *n* represente el número de elementos en una colección, como el número de registros en un conjunto de datos. Aquí hay un ejemplo en el que iteramos una colección de números de tamaño *n*, multiplicamos cada uno por 10 y los sumamos:

$$ \displaystyle\sum_{i=1}^n 10x_{i}$$ 

Usamos Python para ejecutar esta expresión con una lista de cuatro números. Tener en cuenta que en Python (y en la mayoría de los lenguajes de programación en general) normalmente hacemos referencia a los elementos comenzando desde el índice 0, mientras que en matemáticas comenzamos desde el índice 1. Por lo tanto, ajustamos en consecuencia en nuestra iteración comenzando en 0 en la función range()

```python
x = [1, 4, 6, 2]
n = len(x)

summation = sum(10*x[i] for i in range(0,n))
print(summation)
```

Una sumatoria en SymPy se realiza utilizando el operador *Sum()*. En el siguiente código, iteramos *i* desde 1 hasta *n*, multiplicamos cada *i* y los sumamos. Pero luego usamos la función *subs()* para especificar que *n* es 5, lo que entonces iterará y sumará todos los elementos de *i* desde 1 hasta *n*:

```python
from sympy import *

i,n = symbols('i n')

# iterar cada elemento i desde 1 hasta n
# luego, multiplicar y sumar
summation = Sum(2*i,(i,1,n))

# especificar n como 5,
# iterar desde 1 hasta 5
up_to_5 = summation.subs(n, 5)
print(up_to_5.doit())
```

Las sumas no se calculan automáticamente, por lo tanto utilizamos la función *doit()* para ejecutar la expresión.

## Exponentes
Los exponentes representan la multiplicación repetida de un número por sí mismo. La *base* es la variable o valor que estamos elevando a la potencia, y el *exponente* es el número de veces que multiplicamos el valor de la base.

- Cuando multiplicamos exponentes con la misma base, simplemente sumamos los exponentes, lo cual se conoce como la regla del producto.
$$ x^2x^3  = x^{2+3} = x^5 $$
- Cuando dividimos exponentes con la misma base, se restan los exponentes. 
$$ \frac{x^2}{x^5} = \frac{1}{x^3} = x^{-3} $$ 
- Cualquier base elevada a 0 da como resultado 1
$$ x^0 = 1 $$
- Los exponentes fraccionarios son una alternativa de representar raíces
$$ x^{1/2} = \sqrt{x}$$
- El exponente de un exponente se multiplica
$$ {(x^{3})}^2 = x^6$$
- Números irracionales también pueden ser exponentes!
$$ x^{\pi} $$

Podemos resolver estos ejercicios con SymPy

```python
from sympy import *

x = symbols('x')
expr = x**2 / x**5
print(expr)
```

## Logaritmos

Un logaritmo es una función matemática que encuentra un exponente para un número específico y una base determinada. 

$$ a^{x} = b $$
$$ \log_{a}b = x $$

Tiene muchas aplicaciones, desde la medición de terremotos hasta el ajuste del volumen, el logaritmo se encuentra en todas partes. También se utiliza mucho en el aprendizaje automático y la ciencia de datos.

Usemos la función *log* en Python para resolver 
$$ \log_{2}8 = x $$

```python
from math import *

x = log(8,2)
print(x)
```

Cuando no le otorgamos una base, la función *log()* utiliza base 10 por default. En ciencias de datos, por default la base del algoritmo es el número *e*

Los logaritmos también tienen propiedades:
- Multiplicación
$$ \log{(a\times b)} = \log{a} + \log{b} $$
- División
$$ \log{(\frac{a}{b})} = \log{a} - \log{b} $$
- Exponencial
$$ \log{a^n} = n \times \log{a} $$
- Logaritmo de 1 es igual a 0
$$ \log{1} = 0 $$
- Inversa
$$ \log{x^{-1}} = \log{(\frac{1}{x})} = - \log{x} $$

<div class="alert alert-block alert-info">
<b>Nota:</b> Una propiedad del número de Euler es que su función exponencial es su propia derivada, lo cual es conveniente para funciones exponenciales y logarítmicas. Esto, matemáticamente hablando, simplifica varios problemas.
</div>

e es el resultado de la expresión 
$$ (1 + \frac{1}{n})^n $$
a medida que n se hace cada vez más grande, tendiendo a infinito

```python
from math import exp
n = [100, 1000, 10000, 100000, 100000000]
for i in range(0, len(n)):
    euler = (1 + (1 / n[i]))**n[i]
    print("\n n = ", n[i], "\n e =", euler)
print("\n Utilizando función exp(1) e = ", exp(1))
```

## Derivadas
Una derivada indica la pendiente de una función y es útil para medir la tasa de cambio en cualquier punto de una función.

¿Por qué nos importan las derivadas? A menudo se utilizan en el aprendizaje automático y otros algoritmos matemáticos, especialmente con el descenso del gradiente. Cuando la pendiente es 0, eso significa que estamos en el mínimo o máximo de una variable de salida. 

Podemos medir la pendiente en cualquier punto de la curva, y podemos visualizar esto con una línea tangente, una línea recta que "apenas toca" la curva en un punto dado. También proporciona la pendiente en un punto dado. Puedes estimar crudamente una línea tangente en un valor dado de x creando una línea que intersecte ese valor de x y un valor de x vecino muy cercano en la función.

$$ m = \frac{y_{2}-y_{1}} {x_{2} - x_{1}} $$

```python
from sympy import *
x = symbols('x')
f = x**2
p1 = plot(f, xlim = [-5, 5], ylim=[-2,15], show=False)
#  Recta tangente al [2, 4]
xo = 2
x1 = 2
x2 = 2.01

yo = f.subs(x, xo)
y1 = f.subs(x, x1)
y2 = f.subs(x, x2)

m = (y2 - y1) / (x2- x1)

dx_f = m*(x-xo)+ yo 

p2 = plot(dx_f, xlim = [-5, 5], ylim=[-1,15], show=False)
p1.append(p2[0])
p1.show()
```

Cuando nos encontramos con una función exponencial como $ f(x) = x^2 $, la función derivada convertirá el exponente en un multiplicador y luego decrementará el exponente en 1, dejándonos con la derivada $f'(x) = 2x $ La notación $f'(x)$ indica una derivada con respecto a *x*, lo que significa que estamos buscando la pendiente en función del valor de *x*. Entonces, si queremos encontrar la pendiente en *x = 2* y tenemos la función derivada, simplemente sustituimos ese valor de *x* para obtener la pendiente.

$$ f(x) = x^2 $$
$$ \frac{d}{dx}f(x) = \frac{d}{dx}x^2 = 2x $$
$$ \frac{d}{dx}f(2) = 2 (2) = 4 $$

La biblioteca de Python SymPy es gratuita y de código abierto, y es una herramienta útil para calcular derivadas simbólicamente. Por ejemplo:

```python
from sympy import *

x = symbols('x')
f = x**2

dx_f = diff(f)
print(dx_f)
```

```python
# Calcular la derivada o pendiente en x = 2
print(dx_f.subs(x,2))
```

# Derivadas parciales
Estas son derivadas de funciones que tienen múltiples variables de entrada.

En lugar de calcular la pendiente en una función que depende de una sola variable, estamos calculando pendientes con respecto a múltiples variables en varias direcciones. Para cada derivada de una variable dada, consideramos que las otras variables permanecen constantes. 

Tomemos la función $$ f(x,y) = 2x^3 + 3y^3 $$

Las variables *x* e *y* tienen sus propias derivadas, representadas como:

$$ \frac{\partial}{\partial x} $$ 
y 
$$ \frac{\partial}{\partial y} $$  

Estas derivadas indican las pendientes con respecto a cada variable en una superficie multidimensional. Técnicamente, cuando tratamos con múltiples dimensiones, llamamos a estas pendientes "gradientes". 

$$  f(x,y) = 2x^3 + 3y^3 $$
$$ \frac{d}{dx}2x^3 + 3y^3 = 6x^2 $$
$$ \frac{d}{dy}2x^3 + 3y^3 = 9y^2 $$

```python
from sympy import *
from sympy.plotting import plot3d

x,y = symbols('x y')
f = 2*x**3 + 3*y**3

# Calculo de las derivadas parciales para x e y
dx_f = diff(f, x)
dy_f = diff(f, y)

print("dx_f = ", dx_f)
print("dy_f = ", dy_f)

# Grafico de la función f(x,y)
plot3d(f)
```

# La regla de la cadena 

Cuando construyamos una red neuronal, necesitaremos un truco matemático especial llamado **regla de la cadena**. Cuando compongamos las capas de la red neuronal, tendremos que desentrañar las derivadas de cada capa. Pero por ahora, aprendamos la regla de la cadena con un ejemplo algebraico simple. Supongamos que se nos dan dos funciones:
$$ y = x^2 + 1 $$
$$ z = y^3 -2 $$

Observa que estas dos funciones están relacionadas, porque *y* es la variable de salida en la primera función pero es la variable de entrada en la segunda. Esto significa que podemos sustituir *y* de la primera función en la segunda función *z* de la siguiente manera:
$$ z = {(x^2+1)}^2 - 2 $$

Entonces, ¿cuál es la derivada de *z* respecto de *x*?

Veamos que podemos verlo desde otro enfoque, realizando las derivadas de *y* y de *z* de forma separada y luego multiplicarlas:
$$ \frac{dy}{dx}(x^2+1) = 2x $$
$$ \frac{dz}{dy}(y^3-2) = 3y^2 $$
$$ \frac{dz}{dx} = (2x)(3y^2) = 6xy^2 $$

Ahora, sólo queda reemplazar *y* en función de *x*
$$ \frac{dz}{dx} = 6xy^2 = 6x{(x^2+1)^2} $$

Veamos que es lo mismo de esta forma, reemplazando *y* en la función *z* para que quede en función de *x* y derivar por un lado el paréntesis ${(x^2+1)}^3$ y luego dentro del paréntesis $x^2$: 

$$ \frac{dz}{dx}({(x^2+1)^3}-2) $$
$$ = 3{(x^2+1)^2 (2x)} $$
$$ = 6x{(x^2+1)}^2 $$

```python
from sympy import *

z = (x**2 + 1)**3 - 2
dz_dx = diff(z, x)
print("dz_dx = ", dz_dx)
```

Esta es la regla de la cadena, que establece que para una función dada *y* (con variable de entrada *x*) compuesta en otra función *z* (con variable de entrada *y*), podemos encontrar la derivada de *z* con respecto a *x* multiplicando las dos derivadas respectivas juntas:
$$ \frac{dz}{dx} = \frac{dz}{dy} \times \frac{dy}{dx} $$

<div class="alert alert-block alert-info">
<b>Nota:</b> La regla de la cadena es una parte clave del entrenamiento de una red neuronal con los pesos y sesgos adecuados. En lugar de desentrañar la derivada de cada nodo de manera anidada, podemos multiplicar las derivadas a través de cada nodo, lo cual es matemáticamente mucho más fácil.
</div>
